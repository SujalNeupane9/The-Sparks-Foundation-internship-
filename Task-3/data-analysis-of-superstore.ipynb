```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/superstore/SampleSuperstore.csv')
data.head()
```

## Exploratory Data Analysis

```python
data.info()
```

```python
data.describe(include='all').T
```

## Analysis of Sales

```python
# Sales by Ship Mode
sales_by_ship_mode = data.groupby('Ship Mode')['Sales'].sum()
plt.figure(figsize=(8, 6))
sales_by_ship_mode.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Segment
sales_by_segment = data.groupby('Segment')['Sales'].sum()
plt.figure(figsize=(8, 6))
sales_by_segment.plot(kind='bar', color='lightgreen')
plt.title('Total Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by City (top 10 cities by sales)
top_10_cities = data.groupby('City')['Sales'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_10_cities.plot(kind='bar', color='lightblue')
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by State (top 10 states by sales)
top_10_states = data.groupby('State')['Sales'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_10_states.plot(kind='bar', color='lightcoral')
plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Region
sales_by_region = data.groupby('Region')['Sales'].sum()
plt.figure(figsize=(8, 6))
sales_by_region.plot(kind='bar', color='orange')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Category
sales_by_category = data.groupby('Category')['Sales'].sum()
plt.figure(figsize=(8, 6))
sales_by_category.plot(kind='bar', color='lightpink')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sales by Sub-Category (top 10 sub-categories by sales)
top_10_subcategories = data.groupby('Sub-Category')['Sales'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
top_10_subcategories.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Sub-Categories by Sales')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

## Regional Analysis

```python
# Regional analysis: Sales, Profit, and Quantity by Region
region_sales_profit_quantity = data.groupby('Region').agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).reset_index()

# Plotting Sales, Profit, and Quantity by Region
plt.figure(figsize=(12, 5))

# Bar plot for Total Sales by Region
plt.subplot(1, 3, 1)
plt.bar(region_sales_profit_quantity['Region'], region_sales_profit_quantity['Sales'], color='skyblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Bar plot for Total Profit by Region
plt.subplot(1, 3, 2)
plt.bar(region_sales_profit_quantity['Region'], region_sales_profit_quantity['Profit'], color='lightgreen')
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)

# Bar plot for Total Quantity by Region
plt.subplot(1, 3, 3)
plt.bar(region_sales_profit_quantity['Region'], region_sales_profit_quantity['Quantity'], color='salmon')
plt.title('Total Quantity by Region')
plt.xlabel('Region')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Category and Sub-Category analysis

```python
category_subcategory_sales_profit_quantity = data.groupby(['Category', 'Sub-Category']).agg({
    'Sales': 'sum',
    'Profit': 'sum',
    'Quantity': 'sum'
}).reset_index()

# Plotting Sales, Profit, and Quantity by Category and Sub-Category
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Bar plot for Total Sales by Category and Sub-Category
axes[0].barh(category_subcategory_sales_profit_quantity['Sub-Category'], 
              category_subcategory_sales_profit_quantity['Sales'], color='skyblue')
axes[0].set_title('Total Sales by Category and Sub-Category')
axes[0].set_xlabel('Total Sales')
axes[0].set_ylabel('Sub-Category')
axes[0].invert_yaxis()

# Bar plot for Total Profit by Category and Sub-Category
axes[1].barh(category_subcategory_sales_profit_quantity['Sub-Category'], 
              category_subcategory_sales_profit_quantity['Profit'], color='lightgreen')
axes[1].set_title('Total Profit by Category and Sub-Category')
axes[1].set_xlabel('Total Profit')
axes[1].set_ylabel('Sub-Category')
axes[1].invert_yaxis()

# Bar plot for Total Quantity by Category and Sub-Category
axes[2].barh(category_subcategory_sales_profit_quantity['Sub-Category'], 
              category_subcategory_sales_profit_quantity['Quantity'], color='salmon')
axes[2].set_title('Total Quantity by Category and Sub-Category')
axes[2].set_xlabel('Total Quantity')
axes[2].set_ylabel('Sub-Category')
axes[2].invert_yaxis()

plt.tight_layout()
plt.show()
```
