```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

## Question to answer

* What will be predicted score if a student studies for 9.25 hours/day?

```python
url = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

## Analysis of data

```python
df.describe().T
```

```python
## Visualizing the relationship between hours and scores
plt.scatter(df.Hours, df.Scores)
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Hours vs Scores')
plt.show()
```

## Regression Analysis

```python
# X represents the Hours column from the dataframe
X = df.Hours

# y represents the Scores column from the dataframe
y = df.Scores

# Calculate the linear regression parameters
slope, intercept, r, p, std_err = stats.linregress(X, y)

# Define a function for the regression line
def regression(x):
    return slope * x + intercept

# Calculate the dependent variable (predicted scores) using the regression function
dep = list(map(regression, X))
```

```python
sns.scatterplot(x=X, y=y)
plt.plot(X, dep, color='red')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#What will be predicted score if a student studies for 9.25 hours/day?
value = regression(x=9.25)
value
```

* The predicted score will be 92.909 if the student studies for 9.25 hours per day.
